```python
import pandas as pd
ssd=pd.read_csv('DS1_C4_S7_Students_Scores_Data_Practice.csv')
ssd
```

```python
import statistics as st
mean=[]
median=[]
mode=[]
for x in ssd:
    mean.append(ssd[x].mean())
    median.append(ssd[x].median())
    mode.append(st.mode(ssd[x]))
row_head=['mean','median','mode']
column_head=['Statistics','Python','Tableau']
#df
df=pd.DataFrame([mean,median,mode],columns=column_head)
df
```

```python
print(mean)
print(median)
print(mode)
```

```python
df.insert(0,'measure of center tendency',row_head)
df
```

```python
df.to_csv('tudents_Scores_Data_Practice',index=False)
```

# Intepretation:Since this Data is numerical so we give priority to mean and median than mode in this case Python has the most marks as the class averages 81.84 marks in this subject which is higher than other subjects and the median is closer to the mean and higher than other subject with median score of the subject is 82.00 showing python is the subject with most marks scored than other subjects.

```python
mean=[]
variance=[]
coefficent_of_variation=[]
for x in ssd:
    mean.append(ssd[x].mean())
    variance.append(st.pvariance(ssd[x]))
    coefficent_of_variation.append((np.sqrt(st.pvariance(ssd[x]))/ssd[x].mean())*100)
    
```

```python
print(mean)
print(variance)
print(coefficent_of_variation)
```

```python
row_labels=['mean','variance','coefficent_of_variation']
column_head=['Statistics','Python','Tableau']
ddf=pd.DataFrame([mean,variance,coefficent_of_variation],columns=column_head)
ddf
```

```python
for x in ssd:
    lo=[]
    uo=[]
    sub=pd.Series(ssd[x])
    Max=max(sub)
    Min=min(sub)
    Range=Max-Min
    Q1=sub.quantile(0.25)
    Q3=sub.quantile(0.75)
    IQR=Q3-Q1
    UF=Q3+1.5*IQR
    LF=Q1-1.5*IQR
    for i in ssd[x]:
        if (i<LF):
            lo.append(i)
        if (i>UF):
            uo.append(i)
    if (x=="Statistics"):
        Statistics=("Statistics",Min,Max,Range,Q1,Q3,IQR,[UF,LF])
    elif(x=="Python"):
        Python=('Python',Min,Max,Range,Q1,Q3,IQR,[UF,LF])
    else:
        Tableau=('Tableau',Min,Max,Range,Q1,Q3,IQR,UF,LF,[lo,uo])
col_names=['Tableau','Min','Max','Range','Q1','Q3','IQR','UF','LF','OUTLIER']
```

```python
dddf=pd.DataFrame([Statistics,Python,Tableau],columns=col_names)
dddf
```

# Intepretation: Using Inter quartile range we have filtered outliers and for Tableau has most ouliers these are few :[[45, 50, 50, 50, 46, 46], [66, 72, 78, 68, 61...]]
# thus we can say that it is the most skewed data.

# Kurtosis analysis for every subject

```python
import matplotlib.pyplot as plt
import statistics as st
from scipy.stats import norm
```

```python
Statistics=sorted(ssd_Statistics)
Python=sorted(ssd_Python)
Tableau=sorted(ssd_Tableau)
S_mean=st.mean(Statistics)
S_std=st.stdev(Statistics)
P_mean=st.mean(Python)
P_std=st.stdev(Python)
T_mean=st.mean(Tableau)
T_std=st.stdev(Tableau)
plt.figure(figsize=(10,7))
plt.plot(Statistics,norm.pdf(Statistics,S_mean,S_std),label="Statistics")
plt.plot(Python,norm.pdf(Python,P_mean,P_std),label="Python")
plt.plot(Tableau,norm.pdf(Tableau,T_mean,T_std),label="Tableau")
plt.legend()
```

```python
from scipy.stats import kurtosis
print(kurtosis(ssd_Statistics))
print(kurtosis(ssd_Python))
print(kurtosis(ssd_Tableau))
```

# intepretation:
kurtosis of Tableau is 1.44 which is less than 3(kurtosis<3).So Tableau marks distribution shows flat in comparision to lepto-kurtic and with thin tails (Platy-kurtic) that indicates  huge outilers might exist in the distribution. 
Python has -1.27 and stats has  -1.28   both are  Platy-kurtic distribution meaning fewer extreme outliers and values
