```python
import re
ten_col[['Single', 'Multi']] = ten_col['Kerb_Weight'].str.extract(r'(\d+)-?(\d+)?')
ten_col['Single']=pd.to_numeric(ten_col['Single'])
ten_col['Multi']=pd.to_numeric(ten_col['Multi'])
ten_col['Kerb_Weight']=ten_col[['Single', 'Multi']].mean(axis=1)
ten_col=ten_col.drop(columns=['Single', 'Multi'])
ten_col['Kerb_Weight']=ten_col['Kerb_Weight']*2.2
ten_col['Kerb_Weight']
#ten_col['Kerb_Weight']
#ten_col['Kerb_Weight']
```

# 'Gross_Vehicle_Weight'1

```python
ten_col['Gross_Vehicle_Weight']=ten_col['Gross_Vehicle_Weight'].astype(str)
ten_col['Gross_Vehicle_Weight']=ten_col['Gross_Vehicle_Weight'].str.findall(r'\d+').str[0]
ten_col['Gross_Vehicle_Weight']=pd.to_numeric(ten_col['Gross_Vehicle_Weight'], errors='coerce')
ten_col['Gross_Vehicle_Weight']=ten_col['Gross_Vehicle_Weight']*2.2
ten_col['Gross_Vehicle_Weight']
```

# In Kerb_Weight the column where having num+symbols in string and num-num + symbols in string  so we used regular expression and cal mean for especially 'num-num + symbols' this case .for Gross_Vehicle_Weight the data type where mixture of str and num and it had other multiple symbol problems then we decided to convert all to striing as it would suit all data type then we used regular expression to filter out number then we converded it into numeric data type using pandas_to_numeric fn then later mul with 2.2 and obtained pound for it.we also cal pound like this mulp by 2.2 in  Kerb_Weight column
