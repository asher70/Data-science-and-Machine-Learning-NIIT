# Task 1
Extract the table captioned "Player Per Game" from the webpage with the URL below using regular expressions:
URL = "https://www.basketball-
reference.com/leagues/NBA_2021_per_game.html"

```python
import requests
import re
import pandas as pd
from io import StringIO

def extract_player_stats():
    # URL of the basketball statistics
    url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
    
    try:
        # Send HTTP request to get the webpage content
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad status codes
        html_content = response.text
        
        # Regular expression pattern to find the table
        pattern = r'<table.*?id="per_game_stats".*?>(.*?)</table>'
        table_match = re.search(pattern, html_content, re.DOTALL)
        
        if table_match:
            table_html = table_match.group(0)
            
            # Clean the HTML by removing comments
            table_html = re.sub(r'<!--.*?-->', '', table_html, flags=re.DOTALL)
            
            # Use pandas to read the HTML table
            df = pd.read_html(StringIO(table_html))[0]
            
            # Clean up the dataframe
            # Remove rows where 'Rk' is NaN (these are usually header repeats)
            df = df[df['Rk'].notna()]
            
            # Reset the index
            df = df.reset_index(drop=True)
            
            return df
            
        else:
            print("Table not found in the webpage")
            return None
            
    except requests.RequestException as e:
        print(f"Error fetching the webpage: {e}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def main():
    # Extract the table
    stats_df = extract_player_stats()
    
    if stats_df is not None:
        # Display the first few rows
        print("\nFirst few rows of the Player Per Game statistics:")
        print(stats_df.head())
        
        # Save to CSV (optional)
        stats_df.to_csv('nba_player_stats_2021.csv', index=False)
        print("\nData has been saved to 'nba_player_stats_2021.csv'")

if __name__ == "__main__":
    main()
```

# Task 2
Extract the following columns from the dataset to perform the analysis:
a) G
b) MP
c) FG
d) PTS
Check the data of these columns and convert them into
the correct data types for further analysis.

```python
import requests
import re
import pandas as pd
from io import StringIO

def extract_and_process_stats():
    # URL of the basketball statistics
    url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
    
    try:
        # Send HTTP request to get the webpage content
        response = requests.get(url)
        response.raise_for_status()
        html_content = response.text
        
        # Regular expression pattern to find the table
        pattern = r'<table.*?id="per_game_stats".*?>(.*?)</table>'
        table_match = re.search(pattern, html_content, re.DOTALL)
        
        if table_match:
            table_html = table_match.group(0)
            table_html = re.sub(r'<!--.*?-->', '', table_html, flags=re.DOTALL)
            
            # Create initial DataFrame
            df = pd.read_html(StringIO(table_html))[0]
            
            # Clean the DataFrame
            df = df[df['Rk'].notna()].reset_index(drop=True)
            
            # Extract required columns
            selected_columns = ['Player', 'G', 'MP', 'FG', 'PTS']
            df_selected = df[selected_columns]
            
            # Display initial data types
            print("\nInitial Data Types:")
            print(df_selected.dtypes)
            
            # Convert columns to appropriate data types
            df_selected['G'] = pd.to_numeric(df_selected['G'], errors='coerce')
            df_selected['MP'] = pd.to_numeric(df_selected['MP'], errors='coerce')
            df_selected['FG'] = pd.to_numeric(df_selected['FG'], errors='coerce')
            df_selected['PTS'] = pd.to_numeric(df_selected['PTS'], errors='coerce')
            
            # Display converted data types
            print("\nConverted Data Types:")
            print(df_selected.dtypes)
            
            # Basic statistical summary
            print("\nBasic Statistical Summary:")
            print(df_selected.describe())
            
            # Check for any missing values
            print("\nMissing Values Count:")
            print(df_selected.isnull().sum())
            
            return df_selected
            
        else:
            print("Table not found in the webpage")
            return None
            
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def analyze_data(df):
    if df is not None:
        print("\nData Analysis:")
        
        # Games played analysis
        print("\nGames Played Statistics:")
        print(f"Average games played: {df['G'].mean():.2f}")
        print(f"Maximum games played: {df['G'].max()}")
        print(f"Minimum games played: {df['G'].min()}")
        
        # Minutes per game analysis
        print("\nMinutes Per Game Statistics:")
        print(f"Average minutes per game: {df['MP'].mean():.2f}")
        print(f"Maximum minutes per game: {df['MP'].max():.2f}")
        print(f"Minimum minutes per game: {df['MP'].min():.2f}")
        
        # Field goals analysis
        print("\nField Goals Statistics:")
        print(f"Average field goals per game: {df['FG'].mean():.2f}")
        print(f"Maximum field goals per game: {df['FG'].max():.2f}")
        print(f"Minimum field goals per game: {df['FG'].min():.2f}")
        
        # Points analysis
        print("\nPoints Statistics:")
        print(f"Average points per game: {df['PTS'].mean():.2f}")
        print(f"Maximum points per game: {df['PTS'].max():.2f}")
        print(f"Minimum points per game: {df['PTS'].min():.2f}")
        
        # Save processed data
        df.to_csv('processed_nba_stats_2021.csv', index=False)
        print("\nProcessed data has been saved to 'processed_nba_stats_2021.csv'")

def main():
    # Extract and process the data
    processed_df = extract_and_process_stats()
    
    # Analyze the processed data
    if processed_df is not None:
        analyze_data(processed_df)

if __name__ == "__main__":
    main()
```

# Task 3
Calculate mean, median, and kurtosis for the extracted columns.
Use box plots to present the distribution for the extracted columns.

```python
import requests
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
from io import StringIO

def extract_and_process_stats():
    url = "https://www.basketball-reference.com/leagues/NBA_2021_per_game.html"
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        html_content = response.text
        
        pattern = r'<table.*?id="per_game_stats".*?>(.*?)</table>'
        table_match = re.search(pattern, html_content, re.DOTALL)
        
        if table_match:
            table_html = table_match.group(0)
            table_html = re.sub(r'<!--.*?-->', '', table_html, flags=re.DOTALL)
            
            df = pd.read_html(StringIO(table_html))[0]
            df = df[df['Rk'].notna()].reset_index(drop=True)
            
            # Extract columns and create a new DataFrame
            selected_columns = ['Player', 'G', 'MP', 'FG', 'PTS']
            df_selected = df[selected_columns].copy()  # Create a copy to avoid warning
            
            # Convert to numeric
            numeric_columns = ['G', 'MP', 'FG', 'PTS']
            for col in numeric_columns:
                df_selected.loc[:, col] = pd.to_numeric(df_selected[col], errors='coerce')
            
            return df_selected
            
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def calculate_statistics(df):
    numeric_columns = ['G', 'MP', 'FG', 'PTS']
    
    # Create a dictionary to store statistics
    stats = {}
    
    for column in numeric_columns:
        stats[column] = {
            'mean': df[column].mean(),
            'median': df[column].median(),
            'kurtosis': kurtosis(df[column].dropna())
        }
    
    # Display statistics
    print("\nStatistical Analysis:")
    print("-" * 50)
    for column in stats:
        print(f"\n{column} Statistics:")
        print(f"Mean: {stats[column]['mean']:.2f}")
        print(f"Median: {stats[column]['median']:.2f}")
        print(f"Kurtosis: {stats[column]['kurtosis']:.2f}")

def create_box_plots(df):
    numeric_columns = ['G', 'MP', 'FG', 'PTS']
    
    # Set the style using seaborn
    sns.set_style("whitegrid")
    
    # Create individual box plots
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Distribution of NBA Statistics (2021)', fontsize=16, y=1.02)
    
    # Flatten axes for easier iteration
    axes = axes.flatten()
    
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
    
    for i, (column, color) in enumerate(zip(numeric_columns, colors)):
        # Create box plot
        sns.boxplot(y=df[column], ax=axes[i], color=color)
        
        # Add title and labels
        axes[i].set_title(f'{column} Distribution')
        axes[i].set_ylabel('Value')
        
        # Add mean line
        mean_value = df[column].mean()
        axes[i].axhline(y=mean_value, color='red', linestyle='--', alpha=0.5)
        axes[i].text(0.02, mean_value, f'Mean: {mean_value:.2f}', 
                    verticalalignment='bottom')
    
    # Adjust layout
    plt.tight_layout()
    
    # Save the plot
    plt.savefig('nba_stats_boxplots.png', dpi=300, bbox_inches='tight')
    print("\nBox plots have been saved as 'nba_stats_boxplots.png'")
    
    # Display plot
    plt.show()

def create_combined_box_plot(df):
    # Set the style using seaborn
    sns.set_style("whitegrid")
    
    # Create a combined box plot
    plt.figure(figsize=(12, 6))
    
    # Melt the dataframe for easier plotting
    df_melted = df.melt(value_vars=['G', 'MP', 'FG', 'PTS'])
    
    # Create box plot
    sns.boxplot(x='variable', y='value', data=df_melted)
    
    # Customize the plot
    plt.title('Combined Distribution of NBA Statistics (2021)', fontsize=14)
    plt.xlabel('Statistics', fontsize=12)
    plt.ylabel('Value', fontsize=12)
    
    # Save the plot
    plt.savefig('nba_stats_combined_boxplot.png', dpi=300, bbox_inches='tight')
    print("Combined box plot has been saved as 'nba_stats_combined_boxplot.png'")
    
    # Display plot
    plt.show()

def main():
    # Extract and process data
    df = extract_and_process_stats()
    
    if df is not None:
        # Calculate statistics
        calculate_statistics(df)
        
        # Create visualizations
        create_box_plots(df)
        create_combined_box_plot(df)
        
        # Save processed data
        df.to_csv('nba_stats_with_analysis_2021.csv', index=False)
        print("\nProcessed data has been saved to 'nba_stats_with_analysis_2021.csv'")

if __name__ == "__main__":
    main()
```
