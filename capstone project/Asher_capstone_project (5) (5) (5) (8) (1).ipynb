```python
def calculate_residuals(model,features,label):
    predictions=model.predict(features)
    df_results=pd.DataFrame({'Actual':label,'Predicted':predictions})
    df_results['Residuals']=abs(df_results['Actual'])-abs(df_results['Predicted'])
    return df_results
def linear_assumption(model,features,label):
    print('ASSUMPTION1:LINEAR RELATIONSHIP BETWEEN THE TARGET AND THE FEATURES','\n')
    print('CHECKING WITH A SCATTER PLOT OF ACTUAL VS PREDICTED.','PREDICTIONS SHOULD FOLLOW THE DIAGONAL LINE.')
    df_results=calculate_residuals(model,features,label)
    # PLOTTING ACTUAL VS PRREDICTED
    sns.lmplot(x='Actual',y='Predicted',data=df_results,fit_reg=False)
    # PLOTTING THE DIAGONAL LINE
    line_cords=np.arange(df_results.min().min(),df_results.max().max())
    plt.plot(line_cords,line_cords,color='darkorange',linestyle='--')
    plt.title('Actual vs Predicted')
    plt.show()
```

Sports Metric Analysis
In a football competition, each match involves two teamsâ€”one playing at home and the other away. The event organizers are interested in knowing whether playing at home gives any significant scoring advantage.

A sports analytics team collects data from all matches in the season and calculates the average number of goals scored by the home team and the away team in each match.

Sports Metrics Analysis (cont'd)
Conduct a hypothesis test to help the analysts check statistically whether there is a statistically significant difference in the average number of goals per match between home and away teams.

Use a significance level (Î±) = 0.05.

Assume that the goal data are normally distributed. Perform the following tasks to test the hypothesis.


( use away_club_goals sd and use sample size as 50)


# ho:there is not statistically a significant difference in average goals per match between home and away teams using hypothesis testing
# ha:there is a statistically significant difference in average goals per match between home and away teams using hypothesis testing
# so it is a two sided test

```python
# home club 
home=df[['home_club_name','home_club_goals']]
home
# away club
away=df[['away_club_name','away_club_goals']]
away
```

```python
import math
from scipy.stats import norm
```

```python
alpha=0.05#significance level
sample=df['home_club_goals'].sample(n=50,replace=True)
x=sample.mean()#sample mean
mu=df['away_club_goals'].mean()#population mean
s=df['away_club_goals'].std()    #population std
n=len(sample) #sample_size
z_stats=(x-mu)/(s/math.sqrt(n))
p_value=norm.sf(abs(z_stats))
z_critical=norm.ppf(1-(alpha/2))
```

# decision rule:
# if p-value<alpha:reject null hypothesis
# if -zcritical>zstatistics>+z critical:reject null hypothesis

```python
p_value<alpha
```

```python
z_stats>z_critical
```

# so we reject null hypothesis

```python
def iqr_winzoration(df,thresh=1.5):
    df1=df.copy()
    columns=df1.columns
    for col in columns:
        q1=df1[col].quantile(0.25)
        q3=df1[col].quantile(0.75)
        iqr=q3-q1
        lf=q1-thresh*iqr
        uf=q3+thresh*iqr
        df1.loc[df1[col]<lf,col]=lf
        df1.loc[df1[col]>uf,col]=uf
    return df1
num=pd.DataFrame(mik.select_dtypes(exclude='object'))
num=iqr_winzoration(num)
```

# 7.2 At a 0.05 significance level, is there a statistically significant difference in the average goal involvement of away teams between the first half and the second half, given a sample size of 45 and the standard deviation from the first-half goal involvement column?

```python
dsi=df[['goal_involvement','half','away_club_name']]
```

```python
#goal_involvement,'half','away_club_name'
```

# assumption

# ho:no significant difference in average goal involvement between the first half and the second half for away teams
# ha:there is a significant difference in average goal involvement between the first half and the second half for away teams

# two tailes test

```python
First_Half_goal_invo=dsi[(dsi['half']=='First Half')].goal_involvement
Second_Half_goal_invo=dsi[(dsi['half']=='Second Half')].goal_involvement
```

```python
alpha2=0.05#significance level
sample=Second_Half_goal_invo.sample(n=45,replace=True)
x=sample.mean()#sample mean
mu=First_Half_goal_invo.mean()#population mean
s=First_Half_goal_invo.std()    #population std
n=len(sample) #sample_size
z_stats=(x-mu)/(s/math.sqrt(n))
p_value2=norm.sf(abs(z_stats))
z_critical=norm.ppf(1-(alpha/2))
```

# decision rule
# p_value < alpha or -z_critical>z_stats>z_critical
# if true:we reject null hypothesis
# if false:Fail to reject the null hypothesis

```python
p_value2<alpha2
```

```python
z_stats>z_critical
```

# so we reject the  null hypothesis .

# it shows  there is a significant difference in the average goal involvement between the first half away teams and the second half away teams

```python
def iqr_winzorization(df,thresh=1.5):
    df_out=df.copy()
    columns_to_treat=df_out.columns
    for col in columns_to_treat:
        q1=df_out[col].quantile(0.25)
        q3=df_out[col].quantile(0.75)
        iqr=q3-q1
        lower_fence=q1-thresh*iqr
        upper_fence=q3+thresh*iqr
        df_out.loc[df_out[col]<lower_fence,col]=lower_fence
        df_out.loc[df_out[col]>upper_fence,col]=upper_fence
    return df_out
ddf=pd.DataFrame(X.select_dtypes(exclude='category'))
X=iqr_winzorization(X)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1])
plt.show()
```

```python

import matplotlib.pyplot as plt 

# Import the KMeans class from sklearn's cluster module
from sklearn.cluster import KMeans

# Initialize an empty list to store inertia values for each value of k
inertia = []

# Define the range of k values (number of clusters) to try, from 1 to 10
K = range(1, 11)

# Loop through each value of k to perform KMeans clustering
for k in K:
# Create a KMeans instance with 'k' clusters and a fixed random state for reproducibility
    km = KMeans(n_clusters=k, random_state=42)

# Fit the KMeans model on the scaled data 'X_scaled'
    km.fit(X_scaled)

# Append the inertia (sum of squared distances of samples to their closest cluster center) to the list
    inertia.append(km.inertia_)

# Plot the number of clusters (k) against the corresponding inertia values
plt.plot(K, inertia, marker='o') 


plt.xlabel('Number of Clusters (k)')


plt.ylabel('Inertia')


plt.title('Elbow Method For Optimal k')


plt.show()
```

# k=3 is the elbow point(the point from which reduction becomes slower )

```python
Kmeans_model=KMeans(n_clusters=3)
```

```python
# building the model
Kmeans_model.fit(X_scaled)
Kmeans_model
```

```python
cluster_centeroid=Kmeans_model.cluster_centers_
cluster_centeroid
```

```python
labels=Kmeans_model.labels_
labels
```

```python
#using pca(Principal Component Analysis) to reduce feature to 2dimension as we have 4 cluster we need to project it in 2D
from sklearn.decomposition import PCA# importting pcs for reducing dimensionality
#step 1 :reduce high dimensional data to 2d using PCA for vizualization
pca=PCA(n_components=2)# create a pca object to reduce data to 2 dimensionns
X_pca=pca.fit_transform(X_scaled)# apply pca to the scaled data,resulting in 2D projections

# transform the cluster centers from the KMeans model(cluster_centeroid=Kmeans_model.cluster_centers_) into the pcs reduced 2D space
centroids_pca=pca.transform(cluster_centeroid)

# sterp 2:plot the data and cluster centeroids 
plt.figure(figsize=(10,6))
# plot the centeroidds in the PCAA-reduced space
#centroids_pca[:,0],centroids_pca[:,1]:are x and y cordinates
# c=labels=Kmeans_model.labels_:color according to predicted cluster labels
#cmap=colormap
#s:size,alpha:transperency of points
#label ='Data points'
plt.scatter(X_pca[:,0],X_pca[:,1],c=labels,cmap='viridis',s=50,alpha=0.7,label='Data Points')

# plot the centeroids in the PCA -reduced space

#centroids_pca[:,0],
#centroids_pca[:,1]:are x and y cordinates,c=color we have given red and X markers,s:200larger, size(200)   
# edgecolors=black, # black border around centeeroids # label='Centeois'(legends label)
plt.scatter(centroids_pca[:,0],centroids_pca[:,1],c='red',marker='X',s=200,edgecolors='black',label='Centroids')


# add plot details 
plt.title('K-Means Clustering')
plt.xlabel('principal component  1')
plt.ylabel('principal component  2')
plt.legend()# show legend
#plt.grid(True)     
plt.show()
```
