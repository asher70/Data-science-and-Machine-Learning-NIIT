# merging column

```python
result_one = pd.merge(app,game_events,on=['player_id','game_id'])
result_one
```

```python
result_two= pd.merge(result_one,game_lineups,on=['player_id'])
result_two['game_id']=result_two['game_id_x']
```

```python
result_three= pd.merge(result_two,games,on=['game_id'])
result_three
```

```python
result_four= pd.merge(result_three,players,on='player_id')
result_four
```

```python
result_four.info()
```

```python
result_four['game_id_y'].size
```

```python
result_four['competition_id']=result_four['competition_id_x']
result_four['player_name']=result_four['player_name_x']
result_four['date']=result_four['date_x']
result_four['game_id']=result_four['game_id_y']
result_four.drop(['player_name_x','player_name_y','competition_id_x','competition_id_y','date_x','date_y','game_id_x','game_id_y'],axis=1,inplace=True)
```
