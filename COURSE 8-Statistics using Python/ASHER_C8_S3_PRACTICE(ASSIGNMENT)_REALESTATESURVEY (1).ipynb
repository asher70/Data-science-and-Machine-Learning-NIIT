# A) the null and alternate hypothesis



# ASSUME:IS THERE    CHANGE(POSITIVE OR NEGATIVE ) IN THE DATA
# H0=IS THE DATA CORRECT 
# H1= THERE IS CHANGE(POSITIVE OR NEGATIVE ).
# H0=33.88
# H1!=33.88
# IT IS TWO SIDED TEST BECAUSE IT IS EQUAL OR NOT EQUAL SINCE THE INVESTOR WNATS TO KNOW THE WHETHER IT HAS CHANGED OR NOT. 

# B)info as per q

```python
ALPHA=0.05
```

```python
population_mean=33.88
population_sd=1.28
sample_size=49
```

# value calculation

```python
import math
from scipy.stats import norm
```

```python
x=32.18 # sample mean as per q
mu=population_mean
s=population_sd
n=sample_size
alpha=ALPHA
z_statistics=(x-mu)/(s/math.sqrt(n))
print('the z_statistics value is: ',z_statistics)
z_critical=norm.pdf(1-alpha)
print('the z-critical value is:',z_critical)
```

```python
z_critical=norm.pdf(1-alpha)
print('the z-critical value is:',z_critical)
```
