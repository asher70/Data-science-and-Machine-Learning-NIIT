```python
# b)
alpha=0.05
```

```python
# critical value of the test
import math
from scipy.stats import norm
z_critical=norm.pdf(1-alpha)
print('the z-critical value is:',z_critical)
```

```python
# c) 
#details as per q
pop_mean=85 
pop_sd=2.1
sample_size=32 
s_mean=87.5

#converting to symbols as formula:
x=s_mean
mu=pop_mean
s=pop_sd
n=sample_size

#formula
z_statistics=(x-mu)/(s/math.sqrt(n))
print('the z_statistics value is: ',z_statistics)
```

```python
# d)
p_value=norm.sf(abs(z_statistics))
print('the p_value value is : ',p_value)
```

```python
# e) conclusion:
# if -zcritical>z-stata>+zcritical:reject  null hypothesis
z_statistics>z_critical

```

# since as per decision rule as z_statistics>z_critical being true we reject null hypothesis.

```python
# if p_value<alpha :reject  null hypothesis
```

```python
p_value<alpha
```
