```python
# task 2: b)
# simple random sampling:-every item in the population has equal chances of being selected
print(f'simple random sampling mean :{df['price'].sample(n=20,random_state=1).mean()}')

#systematic sampling:the first indvidual is
# randomely selected but others are selected using a fixed sampling interval
s_s=df['price'][::10].mean()
print(f'systematic sampling mean:{s_s}')

# Stratified Sampling:
# population is divided into subgroups(strara) and samples are draen proportionally from each group
p_df=df.groupby(['premium'])
prem=df.premium.unique()
start_asmple_df=pd.DataFrame()
for p in prem:
    daata=p_df.get_group(p)
    #print(daata)
    sample=daata.sample(n=5,random_state=1)
    #print(sample)
    start_asmple_df=pd.concat([start_asmple_df,sample],ignore_index=True)
print(f'Stratified Sampling mean:{start_asmple_df['price'].mean()}')

```

# in different  sampling techniques the mean changes like we got nearly mean to  orginal population mean (2219.58) using systematic sampling mean(2220.65)  using others we got less like Stratified Sampling mean(2098.1) and more  using simple random sampling mean(2465.05)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k=[10,20,30,100,1000,2000,4000,10000]
# create a grid of sunplots 2,4(length of k)
fig,axes=plt.subplots(2,4,figsize=(20,10))
fig.suptitle('distributions of different samples from price',fontsize=18)

# flatten axes for easy looping
axes=axes.flatten()

for i,x in enumerate(k):
    a=df['price'].sample(n=x,random_state=1,replace=True)
    sns.distplot(x=a,kde=True,ax=axes[i])
    axes[i].set_title(f'sample size={x} ')
        
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


```

# as the number of samples increase the distribution of prices increases.
