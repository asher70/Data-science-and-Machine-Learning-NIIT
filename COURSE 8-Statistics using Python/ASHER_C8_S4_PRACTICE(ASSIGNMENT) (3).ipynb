```python
import statistics as st
sample_data=[61,93,75,67,84,95]
s_mean=st.mean(sample_data)
s_sd=st.stdev(sample_data)
n=len(sample_data)
print('sample mean is:',s_mean)
print('sample standard deviation is:',s_sd)
print('sample size is:',n)
p_mean=71
dof=n-1
```

# 

#a)
# ho:mean score for the class is below or same
# ha:mean score for the class is above
# h0:<=71
# ha:>71
# it is one sided right distribution

```python
alpha=0.1
```

```python
import math 
from scipy.stats import t
#b
t_critical=t.ppf(1-alpha,dof)
print('the critical value is:',t_critical)
```

```python
#c)
t_statistic=(s_mean-p_mean)/(s_sd/math.sqrt(n))
print('the test statistic value is:',t_statistic)
p_value=t.sf(abs(t_statistic),dof)
print('the p value is:',p_value)
```

# 

```python
#d)
p_value<alpha
```

```python
t_statistic<t_critical
```

# since critical value is greater than test  statics and
# p value is greater than alpha we do not reject but accept null hypothesis

```python
#ploting the distribution
import matplotlib.pyplot as plt
import numpy as np
upper=s_mean+4*s_sd
lower=s_mean-4*s_sd
x=np.arange(-4,4,.05)
t_vals5=t.pdf(x,5)
#dof=5
plt.plot(x,t_vals5,color='yellow')
plt.legend(['dof=5'])
plt.show()
```

```python
b1=[9,9,9,36,45,48,51,57,69,72,96]
b2=[60,39,39,60,79,96,69,60,75,45,90]
x1=st.mean(b1)
x2=st.mean(b2)
v1=(st.stdev(b1))**2
v2=(st.stdev(b2))**2
n1=len(b1)
n2=len(b2)
dof=n1+n2-2
print('degree of freedom',dof)
print('...............Batch-1...............')
print('mean',x1)
print('standarddeviation:',v1)
print('size:',n1)
print('...............Batch-2...............')
print('mean',x2)
print('standarddeviation:',v2)
print('size:',n2)
```

# 

a)
ho:batch 1 is not equal to batch 2.
ha: both batch are equal.    
ho:b1=b2
ha:b1!=b2
# two sided tail test

```python
# b)
alpha=0.05   #0.1 is significance value since two edge test 0.05
t_critical=t.ppf((1-alpha),dof)
print('critical value:',t_critical)
```

```python
#c)
signal=abs(x1-x2)
noise=math.sqrt(v1*(n1-1)+v2*(n2-1))*math.sqrt(1/n1+1/n2)
t_statistics=signal/noise
print('test statistics:',t_statistics)
p_value=t.sf(abs(t_statistics),dof)*2
print('p value:',p_value)
```

# d)
# As per discussion:
# p_value<alpha:reject null hypothesis
# t_statistics>+tcritical

```python
p_value<alpha
```

```python
t_statistics>t_critical
```

# we have accepted the the null hypothesis
