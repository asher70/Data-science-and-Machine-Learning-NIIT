```python
def check_num_dc(x):
    print(f'Mean:{Employee[x].mean()}')
    print(f'Median:{Employee[x].median()}')
    print(f'The std of a column:{Employee[x].std()}')
    print(f'missing values in the column:{Employee[x].isnull().sum()}')
    if Employee[x].mean()>Employee[x].median():
        print('Skewness:Right skewed')
    elif Employee[x].median()>Employee[x].mean():
        print('Skewness:Left Skewed')
    elif Employee[x].mean()==Employee[x].median():
        print('Skewness:symetric')
    
    #print(f'skewness of a column:{Employee}')
```
