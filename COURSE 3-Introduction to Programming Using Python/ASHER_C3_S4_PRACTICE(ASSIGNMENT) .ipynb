
Task 1: Factorial of a Number
Using "while" loop, write a program to calculate the factorial of a number n. Accept the number n as an input from the user.
Factorial of a number is the product of all consecutive numbers up to that number.
Example:
Input: 5
Processing: 5!= 5x4x3x2x1 Output: 120

```python
n=int(input("ENTER THE NUMBER:"))#5
suma=1
while n>0:  
    suma=suma*(n) 
    n-=1
    
        
print(suma)
# created a variable n for capturing input from user  created a var called suma with init 1 passed a while loop with number greater than 0 and suma mylt 
#n the place holder we get a new suma value then we decrement the n value then we repeat the process until n is zero when it stops then we print suma to get the resultant 
#value of factorial
```

Take input of a five-digit number from the user. Write a program using the "while" loop to calculate the sum of the digits of the number entered by the user.
For example:
If the number entered by the user is 12345, then the output will be 1+2+3+4+5=15.
Note: Use either the "for" loop or the "while" loop to perform this task. Using "for" loops has a simpler approach to this problem.

```python
#for loop
user=(input("ENTER THE NUMBER(5 digit no):"))
t=[int(user[0]),int(user[1]),int(user[2]),int(user[3]),int(user[4])]
emp=0
for x in t:
    emp=emp+x
    
print(emp)
#we stored 5 number in string format using input then we indexed each digit using integer then we stored it in a list and stored in a variable t
#to add we created emp with init 0 then we passedd for loop in list t then we added in emp var each iteration the placeholder with emp var then 
#finally printed emp
```

```python
# while loop
user=(input("ENTER THE NUMBER(5 digit no)"))
t=[int(user[0]),int(user[1]),int(user[2]),int(user[3]),int(user[4])]
suma=0
while t!=[]:
    suma=t[-1]+suma
    del t[-1]
print(suma)
```

Using "while" loop, write a program to display the Fibonacci series up to 7 values. The output would be as follows: 0, 1, 1, 2, 3, 5, 8
Fibonacci series is a sequence of numbers in which
each number is the sum of the two preceding numbers, starting from the third number. The first two numbers are always 0 and 1.
1=0+1
2=1+1
And so on..

```python
lists=[0,1]
count=2
macro=0
while count<7:
    macro=lists[-1]+lists[-2]
    lists.append(macro)
    count+=1

print(lists)
# we created a list stored values 0 and 1 and stored it in lists variable
#we started counting from 2 as init 2 in count var
#since lists had 2 elements already
#we defined macro and initialized as 0 while count was less than 7 as per q
#macro counted lists last and second last element 
# in each iteration we added it to lists using append 
#then we incremented count to 1
# then we printed lists to display op
```

Consider the following data structures: m = {10, 20, 30, 40, 50} # m is a set n = [33, 40, 55, 60, 70] # n is a list

What happened to the elements that were common in the two data structures? Provide explanation.
Note: Use the "add" function to insert elements into the set. You may use either the "for" loop or the "while" loop to perform this task.
