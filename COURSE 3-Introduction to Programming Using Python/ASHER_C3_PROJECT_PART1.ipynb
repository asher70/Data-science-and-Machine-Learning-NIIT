•
display_score(questions, score): Prints a summary of the player's performance, including correct answers, total questions, and percentage of correct responses.
exec_questions(questions): Iterates through a given set of questions, handles user interactions, and keeps track of the score.

```python

def display_score(questions, score):
    percentage_score=(score/(questions))*100
    print("................summary of performance................")
    print(f"  correct answer :{score}")
    print(f"  total_questions :{questions}")
    print(f"  percentage_score :{percentage_score}")
def exec_questions(at,score,questions):
    if at:
        score+=1
    questions+=1
    return score,questions
```
