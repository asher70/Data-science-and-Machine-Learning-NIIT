
Task 3: Looping Over a List
Write a program to count the total number of even and odd numbers in the list defined below.
numlist = list (range (1,30))
The logic should be implemented using a single "for" loop.

```python
numlist = list(range (1,30)) 
count_even=0
count_odd=0
for x in numlist:
    if x%2==0:
        count_even=count_even+1
    else:
        count_odd=count_odd+1

print("The count of even number is:",count_even)
print("The count of odd number is:",count_odd)
#we created two variables to store count of even and odd count then we iterated the list then we applied if the remainder is zero for the placeholder we incremented even count and then applied else clause for odd and printed  both  outside and got the count after all iteration is completed 
```

Select the first 5 even numbers from numlist, store them in a new list, and display the contents of the new list.
numlist = list(range (1,30))
Hint:
el = [] # el is an empty list
Use the append() function to add elements to the new empty list.

```python
el = []                                   # created empty list for storing entries
numlist = list(range (1,30))              # gives a list with 1 to 29 and saved it in numlist variable
count=0                                   # intialized variable count as zero
for x in numlist :                        # x iterated through each element in list numlist
    if x%2==0:                            # to check if the remainder is zero when x modulas 2 == 0 as reminder we do this to gilter even number
        el.append(x)                      #LIST el ADDES x USING append FUNCTION 
        count=count+1                     #counting each iteration
    if count==5: break;                    # we  need to stop at 5 as we need to fetch 5 even numbers we use break to come outside loop

print(el)
        
```
