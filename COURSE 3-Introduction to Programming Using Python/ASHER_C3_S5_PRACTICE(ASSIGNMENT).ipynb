

Task 1: Checking if a word is a palindrome
Create a user-defined function to accept a word from the user and check if the word is a palindrome or not.
Note: Palindromes are the words that read the same backward and forward e.g. madam, mom, dad etc. The word "madam" is a palindrome as it spells the same even after reversing. The word "CAT" is not a
palindrome, since "TAC" would be the reverse of "CAT" and they are not spelled the same.

```python

def palicheck():
    flag=False
    word=input("ENTER THE WORD:")
    if word[::]==word[::-1]:
        print("it is a palindrome")
    else:
        print ("not a palindrome")

    
```

```python
palicheck()
```

Create a user-defined function to accept multiple words from the user and store them in a list. Sort the words in the list in alphabetical order and print them.
Sample Input: [zebra, cat, ant]
Sample Output: [ant, cat, zebra]

```python
def animal():
    a=[]
    while True:
        user=input("ENTER THE WORD( type (stop) to end):") 
        if user=="stop":
            break;
        else:
            a.append(user)
    print(f"stored list :{a}") 
    a.sort()
    print(f"sorted list (alphabetically):{a}")
    
```

```python
animal()
```

Create a user-defined function to allow the user to input 10 numbers and save them in a list. Print the smallest
and the largest numbers from this list.
Sample input:
Data = [3,4,1,0, -31, 20, 40]
Sample output:
Min = -31
Max =40

```python
def it():
    count=0
    a=[]
    while True:
        user=input("ENTER THE NUMBER(TYPE stop TO exit)")
        if user=="stop":break;
            
        elif count<10:
            num=int(user)
            a.append(num)
            #print(count)
            count+=1
        else:
            print("Exceeded limit")
            break;
    print(f"smallest number in the list :{min(a)}")
    print(f"largest number in the list :{max(a)}")
    
        
```

a = ["Andy", "Mandy", "Sandy"]

Write a program to find the common element(s) in the two lists using Lambda expressions.
Sample Input:
a = ["Ant", "Cat", "Mat"]

Sample Output: [Ant]

```python
a = ["Andy", "Mandy", "Sandy"]
b = ["Handy", "Burgundy", "Mandy"] 
#c=a+b
common=list(filter(lambda x :x in a,b))
print(common)
```

c = [12,14,14]
d = [17,16,17]
Write a Python program to add each element of list c with the corresponding element of list d, using the map function and Lambda expression.
The output should be [29,30,31].

```python
c = [12,14,14]
d = [17,16,17] 
op=list(map(lambda x,y:x+y,c,d ))
print(op)
```

Develop a function that accepts the below mentioned date of birth and date of joining of an employee and determines if the employee is over 18 years of age. The function should accept dob and doj as two arguments and return True or False.
dob="12-25-1991" #date of birth
doj= "04-25-2020" #date of joining the company The dates are in MM-DD-YYYY format.
Hint: Only compare the year component in the dates to perform this task. Use the split() function in string. For example, parts_of_date = dob.split("-")

```python
def dobdoj(dob,doj):
    p=dob.split("-")
    w=doj.split("-")
    if int(w[2])-int(p[2])>18:
        return True
    else:
        return False
```

```python
dobdoj("12-25-1991","04-25-2020")
```
