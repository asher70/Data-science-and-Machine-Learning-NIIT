Grade Calculation - Implement a method called calculate_grade() in the Student class that calculates the grade based on the calculated percentage. Grading scale of the institute is given below:
Below 0: Invalid
Above 100: Impossible
80.01 to 100: Grade A
60.01 to 80: Grade B 40.01 to 60: Grade C 20.01 to 40: Grade D 0 to 20: Grade F

```python
class Student():
    def __init__(self,enrollment_ID, student_name, course_name, age, marks,grade):
        self.enrollment_ID=enrollment_ID
        self.student_name=student_name
        self.course_name=course_name
        self.age=age
        self.marks=marks
        self.grade=None
        
    def age_validation(self):
        if self.age<0:
            return 'invalid age'
        elif self.age<3:
            return 'Below age'
        elif self.age>60:
            return 'Over Age'
        else:
            return 'Valid age'
    """Task 3 Marks Validation - Modify the program with a method validate_and_convert_marks() 
    to ensure that the entered marks are between 0 and 44. If the entered marks are within this range,
    convert them into a percentage out of 44 and return the result. Otherwise, return 'Invalid Marks'. 
    """
    
    def validate_and_convert_marks(self):
        if self.marks>=0 and self.marks<=44:
            t=(self.marks/44)*100
            return t
        else:
            return 'Invalid Marks'
    """ Task 4 Grade Calculation - Implement a method called calculate_grade() in the Student
class that calculates the grade based on the calculated percentage.
Grading scale of the institute is given below: Below 0: Invalid Above 100:
Impossible 80.01 to 100: Grade A 60.01 to 80: Grade B 40.01 to 60: Grade C 20.01 to 40: 
Grade D 0 to 20: Grade F"""        
    def calculate_grade(self):
        T=self.validate_and_convert_marks()
        if T<0:
            return 'Invalid'
        elif T>100:
            return 'Impossible'
        elif T>=80.01 and T<=100:
            return 'Grade A'
        elif T>=60.01 and T<=80:
            return 'Grade B'
        elif T>=40.01 and T<=60:
            return 'Grade C'
        elif T>=20.01 and T<=40:
            return 'Grade D'
        elif T>=0 and T<=20:
            return 'Grade F'
```

```python
class Student():
    def __init__(self,enrollment_ID, student_name, course_name, age, marks,grade):
        self.enrollment_ID=enrollment_ID
        self.student_name=student_name
        self.course_name=course_name
        self.age=age
        self.marks=marks
        self.grade=None
        
    def age_validation(self):
        if self.age<0:
            return 'invalid age'
        elif self.age<3:
            return 'Below age'
        elif self.age>60:
            return 'Over Age'
        else:
            return 'Valid age'
    """Task 3 Marks Validation - Modify the program with a method validate_and_convert_marks() 
    to ensure that the entered marks are between 0 and 44. If the entered marks are within this range,
    convert them into a percentage out of 44 and return the result. Otherwise, return 'Invalid Marks'. 
    """
    
    def validate_and_convert_marks(self):
        if self.marks>=0 and self.marks<=44:
            t=(self.marks/44)*100
            return t
        else:
            return 'Invalid Marks'
    """ Task 4 Grade Calculation - Implement a method called calculate_grade() in the Student
class that calculates the grade based on the calculated percentage.
Grading scale of the institute is given below: Below 0: Invalid Above 100:
Impossible 80.01 to 100: Grade A 60.01 to 80: Grade B 40.01 to 60: Grade C 20.01 to 40: 
Grade D 0 to 20: Grade F"""        
    def calculate_grade(self):
        T=self.validate_and_convert_marks()
        if T<0:
            return 'Invalid'
        elif T>100:
            return 'Impossible'
        elif T>=80.01 and T<=100:
            return 'Grade A'
        elif T>=60.01 and T<=80:
            return 'Grade B'
        elif T>=40.01 and T<=60:
            return 'Grade C'
        elif T>=20.01 and T<=40:
            return 'Grade D'
        elif T>=0 and T<=20:
            return 'Grade F'
    """ Task 5 Display Student's Performance - Implement 
a method named display_report() within the Student class, 
responsible for printing the enrollment_ID, student_name, course_name, 
age, marks, and grade. This method should generate and print
the performance report card."""         
    def display_report(self):
        print({'enrollment_ID':self.enrollment_ID, 'student_name':self.student_name, 'course_name':self.course_name,'age':self.age,'marks':self.marks,'grade':self.calculate_grade()})
```
